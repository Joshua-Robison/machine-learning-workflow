# Support Vector Machines: Fit and evaluate a model

Using the Titanic dataset from [this](https://www.kaggle.com/c/titanic/overview) Kaggle competition.

In this section, we will fit and evaluate a simple Support Vector Machines model.

### Read in Data

![Cross-Val](../references/cross_val.png)
![CV](../references/cross_validation.png)

```python
import joblib
import pandas as pd
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
import warnings

warnings.filterwarnings("ignore", category=FutureWarning)
warnings.filterwarnings("ignore", category=DeprecationWarning)

tr_features = pd.read_csv("../data/train_features.csv")
tr_labels = pd.read_csv("../data/train_labels.csv")
```

### Hyperparameter tuning

```python
def print_results(results):
    print("BEST PARAMS: {}\n".format(results.best_params_))
    means = results.cv_results_["mean_test_score"]
    stds = results.cv_results_["std_test_score"]
    for mean, std, params in zip(means, stds, results.cv_results_["params"]):
        print("{} (+/-{}) for {}".format(round(mean, 3), round(std * 2, 3), params))
```

```python
svc = SVC()
parameters = {"kernel": ["rbf", "linear"], "C": [0.01, 1.0, 10]}

cv = GridSearchCV(svc, parameters, cv=5)
cv.fit(tr_features, tr_labels.values.ravel())
```

### Write out pickled model

```python
print_results(cv)
```

```python
cv.best_estimator_
```

```python
joblib.dump(cv.best_estimator_, "../models/SVM_model.pkl")
```
